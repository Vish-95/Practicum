Нужно научиться определять рыночную стоимость объектов недвижимости. Ваша задача — установить параметры. ## Предобработка данных

### Обработка пропусков

Проверим данные на наличие пропусков с помощью метода `isna()` и подсчитаем их количество:

```python
data.isna().sum()
```

Заполним пропуски в параметре `balcony` нулем, так как пропуск скорее всего означает, что балкона нет.
Параметр `floors_total` имеет небольшое количество пропусков, их тоже заполним нулем:

```python
data[['balcony', 'floors_total']] = data[['balcony', 'floors_total']].fillna(0)
```

В данных 49 записей с пропуском в параметре `locality_name`. Предлагаю удалить эти записи, так как восстановление названия населенного пункта вряд ли возможно. 

```python
data = data.dropna(subset=['locality_name']).reset_index(drop=True)
```

Параметр `is_apartment` может влиять на рыночную стоимость объекта недвижимости, но в этом столбце слишком много пропусков - 20924, поэтому оставим его пока что без изменений.

Положим, если в радиусе 3 км есть хотя бы один парк или водоем, то расстояние до ближайшего парка, ближайшего водоема:

```python
data.loc[(data['parks_around3000'] != 0) & (data['parks_nearest'].isna()), 'parks_nearest'] = 3000
data.loc[(data['ponds_around3000'] != 0) & (data['ponds_nearest'].isna()), 'ponds_nearest'] = 3000
```

Можно предположить, что параметры `cityCenters_nearest` и `airports_nearest` можно заполнить их медианой предварительно сгруппировав данные по столбцу `locality_name`:

```python
ls = ['cityCenters_nearest', 'airports_nearest']
for val in ls:
    data[val] = data[val].fillna(
    data.groupby('locality_name')[val].transform('median'))
```

```python
# Приведём столбец с датой размещения объявления к типу datetime
data['first_day_exposition'] = pd.to_datetime(
    data['first_day_exposition'], format='%Y%m%dT%H:%M:%S')

# Построим график распределения пропущенных значений параметра 'days_exposition' 
# относительно дня выставления квартиры на продажу
data[data['days_exposition'].isna()]['first_day_exposition'].value_counts().plot(figsize=(10, 5)
     ).set(xlabel='Дата размещения объявления', ylabel='Количество пропусков', title='Распределения пропущенных значений параметра "days_exposition"');
# plt.xlabel('Дата размещения объявления')
```

 Пропуски в параметре `days_exposition` - сколько дней было размещено объявление (от публикации до снятия), как видно из графика носят неслучайный характер.
 
 Мы посмотрели как распределяются пропуски `days_exposition` в зависимости от даты размещения объявления `first_day_exposition`. Из графика видно, что большая часть пропущенных занчений соответсвует 2019 году и далее число уменьшается. Можно  предположить, что пропуски в столбце `days_exposition` это те объявления которые еще не сняты. Оставим данный столбец без изменений.

Пропущенных данные в столбце `ceiling_height` мы заполним медианой, сгруппировав данные по столбцам название населенного пункта и количество этажей в здании, предполагая, что в одном населенном пункте в зданиях с одинаковой этажностью в среднем одна высота потолка:

```python
# Высота потолков 
data['ceiling_height'] = data['ceiling_height'].fillna(
    data.groupby(['locality_name', 'floors_total'])['ceiling_height'].transform('median'))
```

Рассмотрим параметры, жилая площадь квартиры и площадь кухни:

```python
# Построим диаграмму рассеиния жилой площади в зависимости от общей площади недвижимости
data.plot(x='living_area', y='total_area', kind='scatter', alpha=0.5);

# Диаграмма рассеиния площади кухни в зависимости от общей площади недвижимости
data.plot(x='kitchen_area', y='total_area', kind='scatter', alpha=0.5);
```

Из графиков видно, что общая площадь и жилая площадь имеют положительную корреляцию. Заменим пропущенные значения в столбце `'living_area'` его медианой, предварительно сгруппировав данные по столбцу `'total_area'`:

```python
data['living_area'] = data['living_area'].fillna(
    data.groupby('total_area')['living_area'].transform('median'))

data['kitchen_area'] = data['kitchen_area'].fillna(
    data.groupby('total_area')['kitchen_area'].transform('median'))
```

### Замена типа данных

Для дальнейшего исследования нам необходимо изменить тип данных `object` параметра `'first_day_exposition'` на тип `datetime`:

```python
data['first_day_exposition'] = pd.to_datetime(
    data['first_day_exposition'], format='%Y%m%dT%H:%M:%S')
```

Параметры `'days_exposition'` — сколько дней было размещено объявление, и `'rooms'` - количество комнат, также необходимо привести к другому типу (`'int'`). Для этого сначала заполним оставшиеся пропуски в `'days_exposition'` нулем.

```python
data['days_exposition'] = data['days_exposition'].fillna(0)
data[['days_exposition', 'rooms']] = data[['days_exposition', 'rooms']].astype('int')
```

Мы заменили тип данных в столбце `'first_day_exposition'` на тип `datetime`, воспользовавшись для этого методом `to_datetime`. Параметры `'days_exposition'`, `'rooms'` мы привели к типу `int`.

## Расчёты и добавление результатов в таблицу

Рассчитаем цену квадратного метра и сохраним результат в параметр `'cost_meter'`:

```python
data['cost_meter'] = data['last_price'] / data['total_area']
```

Добавим в таблицу столбцы — день недели, месяц и год публикации объявления:

```python
data['day_exposition'] =  data['first_day_exposition'].dt.weekday
data['month_exposition'] = pd.DatetimeIndex(
    data['first_day_exposition']).month
data['year_exposition'] = pd.DatetimeIndex(
    data['first_day_exposition']).year
```

Добавим классификацию этажа квартиры, варианты — первый, последний, другой:

```python
def classification_floor(row):
    total = row['floors_total']
    floor = row['floor']
    if floor == 1:
        return 'первый'
    elif floor == total:
        return 'последний'
    return 'другой'

data['class_floor'] = data.apply(classification_floor, axis=1)
```

Для исследования и анализа данных добавим следующие отношения:

```python
# Расчет отношения жилой площади к общей
data['living_and_total_area'] = data['living_area'] / data['total_area']

# Расчет отношения площади кухни к общей.
data['kitchen_and_total_area'] = data['kitchen_area'] / data['total_area']
```

Выведем обновленные данные:

Мы рассчитали и добавили в таблицу: стоимость одного квадратного метра, отношение жилой площади к общей и отношения площади кухни к общей. Также мы классифицировали этаж квартиры на — первый, последний и другой.



## Исследовательский анализ данных

С помощью гистограммы распределения исследуем параметры — площадь, цена, число комнат и высота потолков:

```python
characteristic = ['total_area', 'rooms', 'cost_meter', 'ceiling_height']
title_plot = ['Распределение общей площади', "Распределение числа комнат", "Распределение цены (1 кв м)", "Распределение высоты потолков"]
x_label = ['общая площадь', 'число комнат', 'стоимость', 'высота потолков']
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
for index in range(len(characteristic)):
    Q_1, Q_3 = data[characteristic[index]].describe()[[4, 6]]
    IQR = Q_3 - Q_1
    
    i = index%2
    j = index//2
    axes[i, j].hist(data[characteristic[index]], bins=30, range=(Q_1 - 1.5*IQR, Q_3 + 1.5*IQR))
    axes[i, j].set_title(title_plot[index])
    axes[i, j].set_xlabel(x_label[index])
    axes[i, j].set_ylabel('Count')
plt.tight_layout()
plt.show()
```

**Вывод**

Мы построили гистограммы распределения параметров `'rooms'`, `'ceiling_height'`, `'last_price'`, `'total_area'`:

 - У большинства объектов недвижимости указано от 1 до 3 комнат;

 - Высота потолка преимущественно от 2.5 до 2.8 (м);

 - Гистограмма распределения цены за один квадратный метр имеет вид нормального распределения, пиковое значение примерно 80 тыс. р.;

 - Большинство обьектов недвижимости имеют общую площадь квартиры около 45 кв.м.

### Изучение среза данных

Будем считать `rooms` выбросами, если количество комнат меньше 1 и больше 6. Посчитаем сколько таких значений:

```python
print('Количество комнат меньше одной', len(data.query('rooms < 1')))
print('Количество комнат больше шести', len(data.query('rooms > 5')))
```

Таких значений немного, отбросим их:

```python
# Удаление выбросов в параметре  rooms
data = data.query('rooms >= 1 and rooms < 5' )
```

```python
# Аналогично нетипично низкие и высокие потолки признаем выбросами и устраним:
data = data.query('ceiling_height >= 2.5 and ceiling_height < 3.2' )

# срез по общей площади
data = data.query('total_area > 30 and total_area < 120')

# срез по цене 
data = data.query('last_price < 1e+7')
```

### Зависимость стоимости недвижимости от различных факторов

```python
# построим тепловую карту корреляции 
plt.figure(figsize=(10, 7))
ax = sns.heatmap(data.corr(), vmin=-1, vmax=1, center= 0, cmap= 'coolwarm')
```

На тепловой карте видно, что есть высокая корреляция параметров площади, цены и площади, а также сильная взаимозависимость между числом комнат и жилой площадью.

### Анализ изменения цены по степени удаленности от центра

Изучите предложения квартир: для каждой квартиры есть информация о расстоянии до центра. Выделите квартиры в Санкт-Петербурге (locality_name). Ваша задача — выяснить, какая область входит в центр. Создайте столбец с расстоянием до центра в километрах: округлите до целых значений. После этого посчитайте среднюю цену для каждого километра. Постройте график: он должен показывать, как цена зависит от удалённости от центра. Определите границу, где график сильно меняется, — это и будет центральная зона.

```python
# Квартиры в Санкт-Петербурге 
locality_spb = data.query('locality_name == "Санкт-Петербург"').reset_index(drop=True)

# Создаём столбец с расстоянием до центра в километрах и округляем
locality_spb['center_nearest'] = locality_spb['cityCenters_nearest']/1000
locality_spb['center_nearest'] = locality_spb['center_nearest'].astype('int')
```

```python
# Расчет средней цены (за квадратный метр) для каждого километра
locality_spb['mean_cost'] = locality_spb.groupby('center_nearest')['cost_meter'].transform('mean')
locality_spb.head(7)
```

```python
locality_spb.sort_values(by='center_nearest').plot(
    y='mean_cost', x='center_nearest', figsize=(10, 5)).set(ylabel='mean_cost',  title='Зависимость цены от удаленности от центра');
```

**Вывод**

Похоже, что граница центральной части города проходит в 10 км от центра. Имеется только один пик дорогой недвижимости на отдаленном от центра расстоянии. 

### Анализ квартир в центральном сегменте

```python
# Выделим квартиры входящие в зону 10 км от ценра
center_locality_spb = locality_spb.query('center_nearest <= 10')
center_locality_spb.head()
```

Построим гистограммы рапределения для параметров —`'total_area'`, `'mean_cost'`, `'rooms'`, `'ceiling_height'`:

```python
characteristic = ['total_area', 'mean_cost', 'rooms', 'ceiling_height']
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle("Распределение параметров в центральном сегменте", fontsize=18)
for ind in range(len(characteristic)):
    i = ind % 2
    j = int(ind / 2)
    ax=axes[i, j]
    ax.hist(x=center_locality_spb[characteristic[ind]])
    ax.grid(True)
    ax.set_title(f"Распределение {characteristic[ind].upper()}")
    ax.set_xlabel(characteristic[ind])
    ax.set_ylabel('Count')
fig.tight_layout()
plt.show()
```

Рассчитаем коэффициент корреляции и посмотрим есть ли зависимость, в срезе данных по Санкт-Петербургу, цены (`mean_cost`) от факторов: `rooms`, `center_nearest`, `first_day_exposition`, `floor`.


```python
# Корреляция Пирсона цены и рассматриваемых факторов (количество комнат, удалённость от центра, этаж)
characteristic = ['rooms', 'center_nearest', 'floor']
print("Коэффициент корреляции")
for val in characteristic:
    print('mean_cost vs', str(val), center_locality_spb['mean_cost'].corr(center_locality_spb[val]))
print()
```

```python
# Построим графики зависимости цены от числа комнат, удалённость от центра
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
center_locality_spb.pivot_table(index='rooms', values='mean_cost', aggfunc=('median')).plot(
    title='Зависимость цены от числа комнат', ax=axes[0])
center_locality_spb.pivot_table(index='center_nearest', values='mean_cost', aggfunc=('median')).plot(
    title='Зависимость цены от расстояния до центра', ax=axes[1])
plt.tight_layout()
plt.show()
```

```python
spb_out_skirts = locality_spb.query('center_nearest > 10')
parameters = ['total_area', 'rooms', 'cost_meter', 'ceiling_height']

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(15, 12))
plt.subplots_adjust(hspace=0.5)
fig.suptitle("Распределение параметров в центральном и удаленном от центра сегменте", fontsize=18, y=0.95)

for val, ax in zip(parameters, axs.ravel()):

    spb_out_skirts[val].plot(alpha=0.5, kind='hist', label='locality_spb', ax=ax)
    center_locality_spb[val].plot(alpha=0.5, kind='hist', label='center_locality_spb', ax=ax)
    ax.set_title(f"Распределение {val.upper()}")
    ax.set_xlabel(val)
    ax.grid(True)
    ax.legend(loc='upper right')
plt.show()
```
